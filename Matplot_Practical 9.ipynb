#Horizontal Bar Chart

```python
import matplotlib.pyplot as plt_antima
import numpy as np
index = np.arange(5)
values1 = [5,7,3,4,6]
std1 = [0.8,1,0.4,0.9,1.3]
plt_antima.title('A Horizontal Bar Chart')
plt_antima.barh(index,values1,xerr=std1,error_kw={'ecolor':'0.1','capsize':6},alpha=0.7,label='First')
plt_antima.yticks(index+0.4,['A','B','C','D','E'])
plt_antima.legend(loc=5)
```

```python
# Multiseries Stacked Bar Charts


    import matplotlib.pyplot as plt_antima
    import numpy as np
    series1 = np.array([3,4,5,3])
    series2 = np.array([1,2,2,5])
    series3 = np.array([2,3,3,4])
    index = np.arange(4)
    plt_antima.axis([0,4,0,15])
    plt_antima.bar(index,series1,color='r')
    plt_antima.bar(index,series2,color='b',bottom=series1)
    plt_antima.bar(index,series3,color='g',bottom=(series2+series1))
    plt_antima.xticks(index+0.4,['Jan15','Feb15','Mar15','Apr15'])
```

```python
#Pie Charts

import matplotlib.pyplot as plt_antima
labels = ['Nokia','Samsung','Apple','Lumia']
values = [10,30,45,15]
colors = ['yellow','green','red','blue']
plt_antima.pie(values,labels=labels,colors=colors)
plt_antima.axis('equal')
```

```python
import matplotlib.pyplot as plt_antima
labels = ['Nokia','Samsung','Apple','Lumia']
values = [10,30,45,15]
colors = ['yellow','green','red','blue']
explode = [0.3,0,0,0]
plt_antima.title('A Pie Chart')
plt_antima.pie(values,labels=labels,colors=colors,explode=explode,startangle=180)
plt_antima.axis('equal')
```
